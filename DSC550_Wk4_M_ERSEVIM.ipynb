```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv (r'auto-mpg.csv', na_values = [np.nan,'?']) #This converts the "?" into na values - easier to deal with
df.head(10) # defaults to 5 rows, unless specified in '()'
# Import csv and show top 10 rows
```

```python
df=df.drop(['car name'], axis=1) #this drops the car name column
```

```python
df.info() # we confirm that horsepower is a float
```

```python
avghp = df['horsepower'].mean() #calculates the average horsepower for non-nulls
avghp
```

```python
df['horsepower'].fillna(value=avghp, inplace=True) # replaces the NaNs with the avg HP
```

```python
df.sort_values(by = 'horsepower') # Sorting to see highs and lows - no 'NaN' values
```

```python
df.describe() # checking out what values exist for origin
```

```python
df['origin1'] = np.where(df['origin']==1, 1, 0) #reassigning to make dummy variables - creates 3 new columns
df['origin2'] = np.where(df['origin']==2, 1, 0)
df['origin3'] = np.where(df['origin']==3, 1, 0)
```

```python
df.head(15) # show that dummy vars are working
```

```python
df.corr() # corr matrix. Most variables have a strong + or - corr with mpg, except cars with origin = 2
```

As we can see above, there are many features correlated wtih mpg, some of which overlap.
For example, number of cylinders is correlated with horsepower, displacement as well as weight. The higher these values are, the LOWER the mpg is (neg corr). Origin1 has the strongest correlation (pos) and is likely a flag for those cars made in america, since European and particulary Japanese cars have been typically smaller, lighter and more fuel efficient.

```python
corr = df.corr()
# plot the heatmap with colors using seaborn
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

```python
df.plot.scatter(x='weight', y='mpg') #plots the scatter of mpg given weight
```

```python
train=df.sample(frac=0.8,random_state=53) #random state is a seed value
test=df.drop(train.index) #creates an 80/20 split
```

```python
import statsmodels.api as sm # imports alibrary which makes OLS very simple

y = 'mpg'
x = ['weight']

# Fit model on training data with a constant term
model = sm.OLS(train[y], sm.add_constant(train[x])).fit()

model.summary()
```

The results above from fitting the OLS on the training data yields an R-squared value of .67, which means that the regression explains 2/3 of the variance of the weight variable. The F-stat and t-value indicate that the regresion line's slope is extremely unlikely to be due to spurious artifacts in the data.

```python
y = 'mpg'
x = ['weight']

# Fit model on testing data with a constant term
model = sm.OLS(test[y], sm.add_constant(test[x])).fit()

model.summary()
```

The results above from fitting the OLS on the TEST data yields an R-squared value of .785, which means that the regression explains almost 4/5 of the variance of the weight variable. The F-stat and t-value indicate that the regresion line's slope is extremely unlikely to be due to spurious artifacts in the data.

```python
train['sqrt_mpg'] = train['mpg']**0.5

y = 'sqrt_mpg'
x = 'weight'

# Fitting a polynomial regression (transformed mpg to sqrt of mpg)
model = sm.OLS(train[y], sm.add_constant(train[x])).fit()

model.summary()
```

After transforming the mpg field to (mpg)^.5, this regression model effectively becomes a polynomial regression.
The fit to the TRAIN data is .713, a very good fit and the significance stats are very high as well.

```python
train.plot.scatter(x='weight', y='sqrt_mpg') # look at the train scatter data with the transformed mpg field.
```

Above scatter looks more linear, but still slightly concave up, likely due to the other unquantified fields that are corellated with mpg like horsepower.

```python
test['sqrt_mpg'] = test['mpg']**0.5

y = 'sqrt_mpg'
x = 'weight'

# Fitting a polynomial regression (transformed mpg to sqrt of mpg) to TEST data
model = sm.OLS(test[y], sm.add_constant(test[x])).fit()

model.summary()
```

The fit of the (transformed) regression to the TEST data has the highest R-squared score indicating over 82% of the weight variance is explained by the regression.
